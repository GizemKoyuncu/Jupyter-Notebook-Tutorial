# Kernels

Kernel (Çekirdek), kullanıcının kodunu çalıştıran ve kodunu inceleyen bir programdır. IPython, Python kodu için bir çekirdek içerir ve diğer bazı diller için çekirdekler yazılmıştır. Diğer dilleri görmek için [buraya](https://github.com/jupyter/jupyter/wiki/Jupyter-kernels) tıklayınız.

Çekirdek başlangıçta, Jupyter çekirdeğe bir bağlantı dosyası geçirir. Bu, kullanıcı arabirimi ile iletişimin nasıl kurulacağını belirler.

Sırasıyla R,Julia kurulumunu JupyterHub'da gerçekleştirelim.

## IJulia

JupyterHub'a admin kullanıcısı ile girip Julia kurulumunu terminalden gerçekleştirelim.
Terminali açıyoruz ve sırayla kodları çalıştırıyoruz.

> wget https://julialang-s3.julialang.org/bin/linux/x64/1.1/julia-1.1.0-linux-x86_64.tar.gz

![1](https://user-images.githubusercontent.com/61363621/80893052-ab6b1380-8cd7-11ea-9f49-6777c05dbbde.png)

> tar -xvzf julia-1.1.0-linux-x86_64.tar.gz

![2](https://user-images.githubusercontent.com/61363621/80893053-ad34d700-8cd7-11ea-97e6-c095d96a6908.png)

> sudo cp -r julia-1.1.0 /opt/

![3](https://user-images.githubusercontent.com/61363621/80893054-adcd6d80-8cd7-11ea-8c90-1162bc22a4a6.png)

> sudo ln -s /opt/julia-1.1.0/bin/julia /usr/local/bin/julia

![4](https://user-images.githubusercontent.com/61363621/80893056-adcd6d80-8cd7-11ea-811b-7f7f679edb08.png)

> julia

![5](https://user-images.githubusercontent.com/61363621/80893057-ae660400-8cd7-11ea-96c9-2cd0d83a92b6.png)
